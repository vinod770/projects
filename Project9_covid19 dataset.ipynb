```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

```

```python
covid=pd.read_csv('covid19 dataset.csv')
```

```python
covid
```

```python
covid.shape
```

```python
covid.isnull().sum()
```

```python
covid.fillna(covid.mean(),inplace=True)
```

```python
covid
```

```python
covid.drop('Last_Update',axis=1,inplace=True)
```

```python
covid.drop('Country_Region',axis=1,inplace=True)
```

```python
covid.shape
```

```python
covid.drop('Province_State',axis=1,inplace=True)
```

```python
covid.drop('ISO3',axis=1,inplace=True)
```

```python
covid.shape
```

```python
covid
```

```python
x=covid.copy()
x
```

```python
covid_new=covid.drop('Deaths',axis=1)
```

```python
covid_new2=covid_new.drop('Recovered',axis=1)
```

```python
covid_new2
```

```python
x=covid_new2
```

```python
x
```

```python
y=covid.iloc[:,3:5]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=45)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
print(pred)
print("score: ",lr.score(x_train,y_train))
print('Mean absolute error: ',mean_absolute_error(y_test,pred))
print('Mean squared error:  ',mean_squared_error(y_test,pred))
print('Root mean squared error: ',np.sqrt(mean_squared_error(y_test,pred)))
print('r2score: ',r2_score(pred,y_test))
```

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.01)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
rd=Ridge(alpha=0.01)            
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.01)
enr.fit(x_train,y_train)
enrpred=enr.predict(x_test)
print(enr.score(x_train,y_train))

```

```python
#Ridge Regresssor has more score compared with other regressors
```
